```python
import numpy as np
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
from sklearn.cluster import AgglomerativeClustering
import random
import matplotlib.pyplot as plt
```

```python
list = [[random.randint(1, 10), random.randint(1, 10)] for _ in range(10)]
```

```python
list
```

```python
data = np.array(list)
```

```python
plt.Figure(figsize= (6,6))
plt.scatter(data[:,0],data[:,1],c='b',)

for i in range(len(data)):
    plt.annotate(i,xy=[data[i,0],data[i,1]],xytext=[3,3],textcoords='offset points')
```

```python
Z1 = linkage(data,method='single', metric='euclidean',)
```

```python
dendrogram(Z1)
plt.title('single linkage')
plt.tight_layout()
plt.show()
```

```python
f1 = fcluster(Z1,t = 3.5)
print(f1)
```

```python
Z2 = AgglomerativeClustering(n_clusters=1,linkage='single')
Z2.fit_predict(data)
```
